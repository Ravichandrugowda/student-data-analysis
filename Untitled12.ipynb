```python
import matplotlib.pyplot as plt

rice = [57, 50, 47, 30]
wheat = [43, 50, 53, 70]
plt.barh(rice, wheat)
plt.grid(color='green', linestyle=':', linewidth=1.0, axis='y', alpha=0.5)

plt.xlabel('Wheat', fontsize=14)
plt.ylabel('Rice', fontsize=14)
plt.title('The results of Rice/Wheat export', fontsize=20)

plt.show()
```

```python
import matplotlib.pyplot as plt

rice = [57, 50, 47, 30]
wheat = [43, 50, 53, 70]
plt.bar(rice, wheat)
plt.grid(color='green', linestyle=':', linewidth=1.0, axis='y', alpha=0.5)

plt.xlabel('Wheat', fontsize=14)
plt.ylabel('Rice', fontsize=14)
plt.title('The results of Rice/Wheat export', fontsize=20)

plt.show()
```

```python
years = ['2016', '2017', '2018', '2019']
rice = [57, 50, 47, 30]
wheat = [43, 50, 53, 70]
plt.figure(figsize=(10,6))

plt.bar(years, rice, label='Rice')
plt.bar(years, wheat, bottom=rice, label='wheat')

plt.xlabel('Rice', fontsize=14)
plt.ylabel('Wheat', fontsize=14)
plt.title('The results of Rice/Wheat export', fontsize=20)
plt.show()
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors=rng.rand(100)
sizes=1000*rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar();
```
